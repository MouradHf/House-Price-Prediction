```python
# 1. Filtrer les colonnes numériques et gérer les valeurs manquantes
df_numeric = train1.select_dtypes(include=['number']).dropna()

# 2. Calculer la matrice de corrélation
corrmat = df_numeric.corr()

# 3. Seuil dynamique pour les corrélations significatives
threshold = 0.5  # Afficher uniquement les variables ayant une corrélation > threshold avec SalePrice
high_corr_vars = corrmat['SalePrice'][abs(corrmat['SalePrice']) > threshold].index

# 4. Matrice pour les variables sélectionnées
cm = np.corrcoef(df_numeric[high_corr_vars].values.T)

# 5. Heatmap améliorée
sns.set(font_scale=1.2, style="whitegrid")  # Style propre et lisible
plt.figure(figsize=(10, 8))  # Taille ajustée pour une meilleure visibilité

hm = sns.heatmap(
    cm,
    mask=np.triu(np.ones_like(cm, dtype=bool)),  # Masquer la moitié supérieure
    cbar=True,  # Ajouter une barre de couleurs
    annot=True,  # Ajouter les coefficients numériques
    square=True,  # Garder les cases carrées
    fmt='.2f',  # Format des coefficients
    annot_kws={'size': 10},  # Taille des annotations
    yticklabels=high_corr_vars.values,  # Étiquettes pour l'axe Y
    xticklabels=high_corr_vars.values,  # Étiquettes pour l'axe X
    cmap="coolwarm"  # Palette de couleurs
)

# Ajouter un titre
plt.title(
    f'Corrélations fortes (Seuil: {threshold}) avec SalePrice',
    fontsize=16,
    fontweight='bold'
)

plt.tight_layout()  # Ajuster automatiquement les espaces
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler

def find_best_model_using_gridsearchcv(X, y):
    # Normalisation manuelle (recommandée pour la régression linéaire)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Modèles et paramètres
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    
    # Recherche des meilleurs paramètres pour chaque modèle
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        # Utilisation des données normalisées pour les modèles sensibles à l'échelle
        if algo_name == 'linear_regression' or algo_name == 'lasso':
            gs.fit(X_scaled, y)
        else:
            gs.fit(X, y)
            
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    
    # Retourne les résultats sous forme de DataFrame
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

# Appel de la fonction (assurez-vous que X et y sont définis)
results = find_best_model_using_gridsearchcv(X, y)
results
```

```python
from sklearn.pipeline import Pipeline

def find_best_model_using_gridsearchcv(X, y):
    algos = {
        'linear_regression': {
            'model': Pipeline([
                ('scaler', StandardScaler()),
                ('regressor', LinearRegression())
            ]),
            'params': {
                'regressor__fit_intercept': [True, False]
            }
        },
        'lasso': {
            'model': Pipeline([
                ('scaler', StandardScaler()),
                ('regressor', Lasso())
            ]),
            'params': {
                'regressor__alpha': [0.1, 1, 10],
                'regressor__selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error', 'friedman_mse'],
                'splitter': ['best', 'random'],
                'max_depth': [3, 5, 10, None]
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

results = find_best_model_using_gridsearchcv(X, y)
results
```
